```python
df.drop_duplicates(inplace=True)
```

## **Treat outliers**

```python
Q1 = df["todays_gen_act"].quantile(0.25)
Q3 = df["todays_gen_act"].quantile(0.75)
IQR = Q3 - Q1
IQR

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["todays_gen_act"] < lower) |
              (df["todays_gen_act"] > upper)]

print("Outliers count:", outliers.shape[0])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.boxplot(x=df["todays_gen_act"])

plt.title("Box Plot for Actual Power Generation (todays_gen_act)")
plt.xlabel("Actual Generation")

plt.show()
```

A box plot was used to visualize outliers in the todays_gen_act column. The plot showed several extreme values outside the whiskers, indicating the presence of outliers. These outliers are due to variations in power generation across different stations. Since these values represent real operational differences, they were retained for analysis instead of being removed

## **Why Outliers Exist in Dataset**
* Different power stations have different capacities
* Some plants generate very high electricity
* Some plants may generate zero or low power (shutdown/maintenance)

## **Check skewness and apply transformations**

**Skewness** tells whether your data is symmetrical or unevenly distributed.
 * A skewness of 0 means the data is perfectly symmetric.

 * Positive skew (>0) means the distribution has a long tail to the right.

 * Negative skew (<0) means the distribution has a long tail to the left.

## **Summary of Findings**

The analysis of the power generation dataset shows that electricity production varies significantly across different states, sectors, and power stations. While most stations generate moderate levels of power, a few large plants contribute a major portion of total generation. There is a strong relationship between planned and actual generation, but deviations indicate inefficiencies in achieving targets. The data also highlights that not all power stations utilize their full capacity, suggesting scope for improvement in operational efficiency.

## **Key Insights**

* A small number of power stations contribute a large share of total generation, while many stations produce relatively low output.

* Although planned and actual generation are positively correlated, actual output often falls short of planned values, indicating inefficiencies.

* Higher capacity plants generally produce more electricity, but many are not operating at full capacity.

* Some stations show extremely high or low generation values, reflecting differences in plant size or operational status.

* Power generation varies across sectors, showing that certain sectors (e.g., thermal or private) contribute more efficiently than others.

## **Recommendations**

* Improve monitoring of plants where actual generation is consistently lower than planned.

* Optimize utilization of high-capacity plants to maximize output.

* Investigate low-performing stations and reduce downtime or maintenance delays.

* Use historical trends to improve future generation planning.

* Balance power distribution by investing in underperforming regions.

# Conclusion

The analysis shows that power generation varies across states and stations, with a strong relationship between planned and actual generation. However, there are gaps indicating inefficiencies and underutilization of capacity. Improving monitoring and optimization can help increase overall power generation efficiency.
